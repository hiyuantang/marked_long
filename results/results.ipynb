```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_results(path):
    results = []
    with open(path) as text:
        for line in text:
            if 'val_acc1' in line:
                i = line.index(':', -25)
                results.append(float(line[i+2:-2]))
    return results
```

```python
L3eidiv = get_results('./3Leidiv.txt')
#L5eidiv = get_results('./5Leidiv.txt')
#L7eidiv = get_results('./7Leidiv.txt')
L3 = get_results('./3L.txt')
L5 = get_results('./5L.txt')
L7 = get_results('./7L.txt')
```

```python
plt.style.use('seaborn-darkgrid')
dpi = 96
plt.figure(figsize=(800/dpi, 800/dpi), dpi=dpi)
palette = plt.get_cmap('Set1')
plt.plot(L3eidiv, marker='', color=palette(0), linewidth=1, alpha=0.9, label='3Leidiv')
#plt.plot(L5eidiv, marker='', color=palette(1), linewidth=1, alpha=0.9, label='5Leidiv')
#plt.plot(L7eidiv, marker='', color=palette(2), linewidth=1, alpha=0.9, label='7Leidiv')
plt.plot(L3, marker='', color=palette(3), linewidth=1, alpha=0.9, label='3L')
plt.plot(L5, marker='', color=palette(4), linewidth=1, alpha=0.9, label='5L')
plt.plot(L7, marker='', color=palette(6), linewidth=1, alpha=0.9, label='7L')

plt.legend(loc=2, ncol=2)
plt.title("EIDivnorm Performance Comparison", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel('epochs')
plt.ylabel('Validation Accuracy')

plt.show()
```
