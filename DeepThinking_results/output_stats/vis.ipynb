```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import torch
from turtle import color
```

```python
from cProfile import label


def get_val_results(path):
    results = [] 
    with open(path) as text:
        for line in text:
            if 'val_acc' in line:
                i = line.index(':', -10)
                results.append(float(line[i+2:-2]))
    return results

def simple_vis(acc_list, name_list, title='Validation Accuracy Comparison'):
    plt.style.use('seaborn-darkgrid')
    dpi = 150
    linewidth=1.5
    legend_loc = 'upper left'
    fig, ax = plt.subplots(1,1,sharex=True, sharey=True, 
                            figsize=(800/dpi, 800/dpi), dpi=dpi)
    palette = plt.get_cmap('Set1')
    for i in range(len(name_list)):
        ax.plot(acc_list[i,:], label=name_list[i])
    ax.set_ylim(0,100)
    ax.legend(loc=legend_loc)
    ax.set_title(title)
    plt.show()

def get_mean_std(*args):
    merge=[]
    for acc in args:
        merge=np.concatenate((merge, acc))
    merge=merge.reshape(-1,5)
    return np.mean(merge,0), np.std(merge,0)
```
